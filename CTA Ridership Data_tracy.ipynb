```python
#RED 
#1.binning the data into 3 year ranges
red["year"]=red["month_beginning"].str[-4 :]
red["year"]=pd.to_numeric(red["year"])
bins=[2000,2006.5,2012.5,2017.5]
group_names=["2001-2006","2007-2012","2012-2017"]
red["YearBins"]=pd.cut(red["year"], bins, labels=group_names)

#2. Finding the total ridership of each station, grouped by year ranges
red_bystation=red.groupby(["YearBins","stationame"]).sum()
red_bystation["BinTotal"]=red_bystation["monthtotal"]

#3. Dropping all other useless columns 
red_bystation.drop("Unnamed: 0", axis=1, inplace=True)
red_bystation.drop("MAP_ID", axis=1, inplace=True)
red_bystation.drop("avg_weekday_rides", axis=1, inplace=True)
red_bystation.drop("avg_saturday_rides", axis=1, inplace=True)
red_bystation.drop("avg_sunday-holiday_rides", axis=1, inplace=True)
red_bystation.drop("monthtotal", axis=1, inplace=True)
red_bystation.drop("year", axis=1, inplace=True)

#4.Finding out stations with the most riders, grouped by year ranges
red_bystation=red_bystation.sort_values(["YearBins", "BinTotal"], ascending=[True,False])


red_bystation_overyear=red_bystation.unstack(0)

change=[]
change=red_bystation_overyear["BinTotal"]["2012-2017"]-red_bystation_overyear["BinTotal"]["2001-2006"]
change.sort_values(ascending=False)

total=[]
total=red_bystation_overyear["BinTotal"]["2012-2017"]+red_bystation_overyear["BinTotal"]["2007-2012"]+red_bystation_overyear["BinTotal"]["2001-2006"]
total.sort_values(ascending=False)                                                                                     
                                                                                      
summary=pd.DataFrame({"change": change, "total": total})
red_combined=pd.merge(red_bystation_overyear, summary,on="stationame", how="outer")
red_combined.sort_values("change", ascending=False)                                                                            


red_combined.sum()
red_change=[341332796, 377529533,333383814]

red_bystation.sum()




                                                                               
```
