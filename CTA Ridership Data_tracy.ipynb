```python
totalriders_byyear=[]
cta2001=cta_data_df.head(365)
cta2001_totalriders=cta2001["total_rides"].sum()
totalriders_byyear.append(cta2001_totalriders)

cta2002=cta_data_df[365:730]
cta2002_totalriders=cta2002["total_rides"].sum()
totalriders_byyear.append(cta2002_totalriders)


cta2003=cta_data_df[730:1095]
cta2003_totalriders=cta2003["total_rides"].sum()
totalriders_byyear.append(cta2003_totalriders)

cta2004=cta_data_df[1095:1461]
cta2004_totalriders=cta2004["total_rides"].sum()
totalriders_byyear.append(cta2004_totalriders)


cta2005=cta_data_df[1461:1826]
cta2005_totalriders=cta2005["total_rides"].sum()
totalriders_byyear.append(cta2005_totalriders)

cta2006=cta_data_df[1826:2191]
cta2006_totalriders=cta2006["total_rides"].sum()
totalriders_byyear.append(cta2006_totalriders)

cta2007=cta_data_df[2191:2556]
cta2007_totalriders=cta2007["total_rides"].sum()
totalriders_byyear.append(cta2007_totalriders)

cta2008=cta_data_df[2556:2922]
cta2008_totalriders=cta2008["total_rides"].sum()
totalriders_byyear.append(cta2008_totalriders)

cta2009=cta_data_df[2922:3287]
cta2009_totalriders=cta2009["total_rides"].sum()
totalriders_byyear.append(cta2009_totalriders)

cta2010=cta_data_df[3287:3652]
cta2010_totalriders=cta2010["total_rides"].sum()
totalriders_byyear.append(cta2010_totalriders)

cta2011=cta_data_df[3652:4048]
cta2011_totalriders=cta2011["total_rides"].sum()
totalriders_byyear.append(cta2011_totalriders)


cta2012=cta_data_df[4048:4414]
cta2012_totalriders=cta2012["total_rides"].sum()
totalriders_byyear.append(cta2012_totalriders)


cta2013=cta_data_df[4414:4779]
cta2013_totalriders=cta2013["total_rides"].sum()
totalriders_byyear.append(cta2013_totalriders)

cta2014=cta_data_df[4779:5175]
cta2014_totalriders=cta2014["total_rides"].sum()
totalriders_byyear.append(cta2014_totalriders)


cta2015=cta_data_df[5175:5540]
cta2015_totalriders=cta2015["total_rides"].sum()
totalriders_byyear.append(cta2015_totalriders)

cta2016=cta_data_df[5540:5906]
cta2016_totalriders=cta2016["total_rides"].sum()
totalriders_byyear.append(cta2016_totalriders)

cta2017=cta_data_df[5906:6271]
cta2017_totalriders=cta2017["total_rides"].sum()
totalriders_byyear.append(cta2017_totalriders)

x_axis=np.arange(0,len(totalriders_byyear))
x_years=["2001", "2002", "2003","2004", "2005", "2006", "2007", "2008", "2009", "2010", "2011", "2012","2013", "2014", "2015", "2016", "2017"]
plt.plot(x_axis, totalriders_byyear)
plt.xticks(x_axis,x_years, rotation ="vertical")
plt.title("Total CTA Ridership Over the Years")


```
